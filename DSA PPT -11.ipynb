## Q1. Make the String Great


```python
def make_good(s):
    
    stack=[]
    
    for i in s:
        if stack and abs(ord(stack[-1]) - ord(i)==32):
            stack.pop()
        else:
            stack.append(i)
            
    return "".join(stack)

s="leEeetcode"
make_good(s)
```

## Q2. Remove all adjacent duplicates in a String

```python
def remove_duplicates(s):
    stack=[""]
    for i in s:
        if stack[-1]==i:
            stack.pop()
        else:
            stack.append(i)
    return "".join(stack)

s = "abbaca"
remove_duplicates(s)
```

## Q4. Time Needed to buy tickets

```python
def time_required_to_buy(tickets,k):
    
    num_sec=0
    
    while tickets[k]>0:
        for i in range(len(tickets)):
            if tickets[i]>0 and tickets[k]>0:
                tickets[i]-=1
                num_sec+=1
    return num_sec

tickets = [2,3,2]
k = 2
time_required_to_buy(tickets,k)
```

## Q6. Largest Rectangle in Histogram

```python
def largestRectangleArea(bars):
    st, res = [], 0
    for bar in bars + [-1]: # add -1 to have an additional iteration
        step = 0
        while st and st[-1][1] >= bar:
            w, h = st.pop()
            step += w
            res = max(res, step * h)

        st.append((step + 1, bar))

    return res

heights=[2,1,5,6,2,3]
largestRectangleArea(heights)
```

## Q7. Sliding Window Maximum

```python
from collections import deque
def maxSlidingWindow(nums, k):

        q, res = deque(), [] 
        for r in range(len(nums)):
            while q and nums[q[-1]] < nums[r]:
                q.pop()
            q.append(r)
            if r+1 < k: continue
            if q[0] < r+1-k:
                q.popleft()
            res.append(nums[q[0]])

        return res
    
nums=[1,3,-1,-3,5,3,6,7]
k=3
maxSlidingWindow(nums,k)
```

## Q8. Design a circular queue data structure with the following operations: enQueue, deQueue, Front, Rear, and isEmpt

```python
class CircularQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size
        self.front = self.rear = -1

    def is_empty(self):
        return self.front == -1

    def is_full(self):
        return (self.rear + 1) % self.size == self.front

    def enQueue(self, item):
        if self.is_full():
            print("Queue is full. Cannot enQueue.")
            return

        if self.is_empty():
            self.front = self.rear = 0
        else:
            self.rear = (self.rear + 1) % self.size

        self.queue[self.rear] = item

    def deQueue(self):
        if self.is_empty():
            print("Queue is empty. Cannot deQueue.")
            return None

        item = self.queue[self.front]

        if self.front == self.rear:
            self.front = self.rear = -1
        else:
            self.front = (self.front + 1) % self.size

        return item

    def Front(self):
        if self.is_empty():
            return None
        return self.queue[self.front]

    def Rear(self):
        if self.is_empty():
            return None
        return self.queue[self.rear]


if __name__ == "__main__":
    cq = CircularQueue(5)
    print(cq.is_empty())  

    cq.enQueue(1)
    cq.enQueue(2)
    cq.enQueue(3)
    cq.enQueue(4)
    cq.enQueue(5)

    print(cq.is_full()) 

    print(cq.Front())  
    print(cq.Rear())  

    cq.deQueue()
    cq.enQueue(6)

    print(cq.Front())  
    print(cq.Rear())   
```
